```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from ast import literal_eval

df = pd.read_csv('../data_samples/data_jobs.csv')

# Data Cleaning
df['job_posted_date'] = pd.to_datetime(df['job_posted_date']).dt.normalize()

def take_list(str_list):
    if pd.notna(str_list):
        return literal_eval(str_list)
    return str_list

df['job_skills'] = df['job_skills'].apply(take_list)

df
```

```python
# Filter the city you want
df_city = df[(df['job_country'] == 'United States') & (df['job_title_short'] == 'Data Analyst')].reset_index(drop=True)

df_city
```

```python
# Add month name and number
df_city['job_posted_month_name'] = df_city['job_posted_date'].dt.month_name()
df_city['job_posted_month_number'] = df_city['job_posted_date'].dt.month
```

```python
# Explode the skills
df_skills = df_city.explode('job_skills')
df_skills = df_skills.sort_values('job_posted_month_number')
```

```python
# Pivot the data by month
df_pivoted = df_skills.pivot_table(index='job_skills', columns='job_posted_month_name', aggfunc='size', sort=False)

# Drop na values
df_pivoted = df_pivoted.dropna()

# Adding total appearnce of the skill throw the year
df_pivoted['total'] = df_pivoted.sum(axis=1)

# Sort it by total
df_pivoted = df_pivoted.sort_values(by='total', ascending=False)

# Select top 5
df_pivoted = df_pivoted.head(5)

# Drop the total
df_pivoted = df_pivoted.drop(axis=1, labels='total')

df_pivoted
```

```python
# Now to plot it we need to make it long format
df_plot = df_pivoted.reset_index().melt(
    id_vars="job_skills", 
    var_name="month", 
    value_name="count"
)

ax = sns.lineplot(data=df_plot, x='month', y='count', hue="job_skills")
sns.despine()


for label in ax.get_xticklabels():
    label.set_rotation(90)

ax.set_ylabel('Count of Jobs')
ax.set_xlabel('')
ax.legend(title='Skills')
ax.set_title('Skills Trending By Jobs Count')
```

```python
# To get the percentages we need to know how many job we have per month (Data Anaylsis)

jobs_mo = df_city.groupby('job_posted_month_name').agg('size').to_frame('count')

# Get the count of the skill divide it by the total jobs count 
# of data anaylst in that month
def find_percentage(row):
    return round(row['count'] / jobs_mo.loc[row['month']]['count'] * 100, 2)


df_plot['percentage'] = df_plot.apply(axis=1, func=find_percentage)
```

```python
# Let's plot
ax = sns.lineplot(data=df_plot, x='month', y='percentage', hue="job_skills")
sns.despine()
for label in ax.get_xticklabels():
    label.set_rotation(90)

# Move the legend outside the plot
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
ax.set_title('Trending Top Skills for Data Anaysis in The USA')
ax.set_xlabel('')
ax.set_ylabel('Likelihood in Job Posting')

# Make the axis cleaner
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda y, _: f'{int(y)}%'))
```
